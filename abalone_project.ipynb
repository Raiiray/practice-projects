```python
import statsmodels.api as sm
import scipy
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x = abalone.iloc[:,:-1]
x
```

```python
y = abalone.iloc[:,-1]
y.shape
```

```python
def calc_vif(x):
    vif = pd.DataFrame()
    vif["variables"] = x.columns
    vif["V_I_F"] = [variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
calc_vif(x)
```

```python
x.drop(["Length","Shucked weight"],axis=1,inplace=True)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
maxr2_score=0
maxrs=0
for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(xt,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    r2=r2_score(np.array(y_test),np.array(predlr))
    if r2 > maxr2_score:
        maxr2_score=r2
        maxrs=i
print("best accuracy score is",maxr2_score,"on random state",maxrs)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report
```

```python
maxacc=0
maxrs=0
for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(xt,y,test_size=0.2,random_state=i)
    dtc=DecisionTreeClassifier()
    dtc.fit(x_train,y_train)
    pred=dtc.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc > maxacc:
        maxacc=acc
        maxrs=i
print("best accuracy score is",maxacc,"on random state",maxrs)
```

```python
maxacc=0
maxrs=0
for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(xt,y,test_size=0.2,random_state=i)
    svc=SVC()
    svc.fit(x_train,y_train)
    predsvc=dtc.predict(x_test)
    acc=accuracy_score(y_test,predsvc)
    if acc > maxacc:
        maxacc=acc
        maxrs=i
print("best accuracy score is",maxacc,"on random state",maxrs)
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(svc,x_train,y_train,cv=5)
score.mean()
```

```python
maxacc=0
maxrs=0
from sklearn.ensemble import RandomForestClassifier
for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(xt,y,test_size=0.2,random_state=i)
    rfc=RandomForestClassifier(n_estimators=200)
    rfc.fit(x_train,y_train)
    predrfc=rfc.predict(x_test)
    acc=accuracy_score(y_test,predrfc)
    if acc > maxacc:
        maxacc=acc
        maxrs=i
print("best accuracy score is",maxacc,"on random state",maxrs)
```
